```python
import torch
from torch import nn
import numpy as np
```

```python
def sigmoid(z):
  return 1 / (1 + torch.exp(-z))
```

```python
x = torch.tensor([
    [0.333, 0.5, 5.0],
    [0.666, 0.25, 10.0],
    [0.999, 0.1, 1.0],
    [0.1, 0.1, 0.2]
])
```

```python
w1 = torch.tensor([
    [-1.5, -0.5],
    [0.0, 1.0],
    [1.5, 3.0]
])

w2 = torch.tensor([
    [2.5, 0.5, -1.5],
    [3.5, 1.5, 1.5]
])

b1 = torch.tensor([0., 0.], requires_grad=True)
b2 = torch.tensor([0., 0., 0.])
```

```python
y = torch.tensor([1, 2, 0, 1])
y_hat = sigmoid(x @ w1 + b1) @ w2 + b2
```

```python
y_hat
```

```python
loss = torch.nn.CrossEntropyLoss(reduction='mean')
output = loss(y_hat, y)
```

```python
output.backward()
```

```python
import numpy as np
import torch
from torch import nn


class MLPClassifier(nn.Module):
    def __init__(self, input_size, output_size=1):
        super(MLPClassifier, self).__init__()

        self.flatten = nn.Flatten()

        self.input_layer = nn.Linear(input_size, 16)
        self.input_activation = nn.ReLU()

        self.hidden_layer = nn.Linear(16, 16)
        self.hidden_activation = nn.ReLU()

        self.output = nn.Linear(16, output_size)
        self.output_activation = nn.ReLU()

    def forward(self, x):

        x = self.flatten(x)

        x = self.input_layer(x)
        x = self.input_activation(x)

        x = self.hidden_layer(x)
        x = self.hidden_activation(x)

        x = self.output(x)
        x = self.output_activation(x)

        return x
```

```python
model = MLPClassifier(input_size=128, output_size=10)
input_data = torch.tensor(np.random.rand(1, 128)).float()
model(input_data)
```
