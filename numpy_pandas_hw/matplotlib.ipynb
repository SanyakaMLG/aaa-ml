```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 8))
f.suptitle('Plot examples', fontsize=20)

x, y = zip(*np.random.multivariate_normal(mean=(20, 10), cov=((1, 0.7), (0.7, 1)), size=1000))
ax1.scatter(x, y, edgecolor='black')
ax1.set_title('Scatter plot example')

x = np.random.geometric(p=0.1, size=3000)
ax2.hist(x, bins=40, color='red')
ax2.set_title('Hist plot example')

x = np.linspace(0.001, 6)
log = lambda x: math.log(x)
ax3.plot(x, np.log(x))
ax3.plot(x, np.log(2*x))
ax3.legend(['log(x)', 'log(2x)'])
ax3.set_title('Line plot example')

x = np.random.normal(loc=5, size=14)
y = np.arange('2023-01-01', '2023-01-15', dtype='datetime64[D]')
ax4.bar(y, x)
ax4.set_xticks(y)
ax4.set_xticklabels(y, rotation=90)
ax4.set_title('Bar plot example')

plt.show()
```
